```python
import pandas as pd
```

```python
last_season = pd.read_csv('lastseason.csv', skiprows=1)
this_season = pd.read_csv('thisseason.csv', skiprows=1)
this_season = this_season.loc[this_season["FG Att"] > 2]
returners = last_season[last_season['Player'].isin(this_season['Player'])]
```

```python
import matplotlib.pyplot as plt
def player_bar_graph(df, stat):
    df = df.sort_values("FG Att", ascending=False)
    plt.bar(df['Player'], df[stat])
    plt.xticks(rotation=30, ha='right')
    plt.ylabel(stat)
```

```python
plt.title("This Season")
player_bar_graph(this_season, "FG Att")
```

```python
plt.title("This Season")
player_bar_graph(this_season, "eFG%")
```

```python
plt.title("Last Season Returners")
player_bar_graph(returners, "eFG%")
```

```python
plt.title("Returners (Last )")
player_bar_graph(this_season, "FG Att")
```
